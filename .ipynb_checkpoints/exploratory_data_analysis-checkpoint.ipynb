```python
# Selecting desired year
def year(df, year):
    """
    Able to select rows with
    desired year
    """
    df.loc[df["year"] == year]

freedom_df = year(freedom_df, 2017)
```

```python
# Creating a dataframe that shows columns of the year 2010
freedom_df2 = year(freedom_df, 2008)

freedom_df2.head(500)
```

Missing values in this dataset are filled with a dash(-)

```python
# Finding rows with missing values
missing_values = freedom_df.loc[freedom_df.pf_association == "-"]  # All nulls are located in pf_association column
len(missing_values)  # 26/162 16.04% of our data is cut out
```

```python
# Dropping missing values
freedom_df = freedom_df.drop(index=missing_values.index)
```

```python
freedom_df.shape
```

### Change data types

```python
# Changing all objects to floats
freedom_df = freedom_df.astype(float)
freedom_df.dtypes
```

```python
# Distribution of the dependent variable, hf_score
sns.distplot(freedom_df.hf_score)

print("hf_score mean:", freedom_df.hf_score.mean())  # Mean of hf_score
print("hf_score sigma:", freedom_df.hf_score.std())  # Standard deviation of hf_score
```

**Observation**

Majority of the results in hf_score is around 6.95,
and the values are tightly clustered around the mean
