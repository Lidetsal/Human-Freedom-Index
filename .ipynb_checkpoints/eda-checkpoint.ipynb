## Data Conclusion

Our data consist of continuous values.

We want the target to be "quality" which is ordinal categorical.

We can convert it using One-Hot-Encoding

### Creating subsets

```python
# Creating Target and Predictors
y = wine_df["quality"]  # Target. The variable that we are trying to predict
"""
The quality has ordinal categorial values of 3-7
"""
X = wine_df.drop(columns=["quality"], axis=1)  # Predictor. Features that will aid in the prediction of the target

# Creating Training and Testing subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 50, test_size = 0.2)
display(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```
